```python
import networkx as nx
```

A graph data Structure containing the k-Nearest Neighbours and the clusters and the required functions

```python
class Graph:
    def __init__(self) -> None:
        Data = open(r"cluster.csv", "r")
        next(Data, None)  # skip the first line in the input file
        Graphtype = nx.DiGraph()
        self.G = nx.parse_edgelist(Data, delimiter=',', create_using=Graphtype,nodetype=int, data=(('Amount', int),))
        self.knn = {}
        self.cluster = {}
    
    #Function to create Single clusters from where the clustering will start
    def create_cluster(self):
        for i in self.G:
            self.cluster[i] = set([i])
    
    
    #Function to find the k nearest neighbours
    def knn_create(self,k):
        for i in self.G:
            self.knn[i] = set([x for x,y in sorted(self.G[i].items(),key=lambda x:x[1]['Amount'])][:k])

    
    #Function which will perform the shared nearest neighbour graph clustering
    def sNN_cluster(self,k,kt):
        
        #finding the knn from the data
        self.knn_create(k)
        
        #creating starting clusters
        self.create_cluster()
        
        #iterating over each of the vertices
        for s in self.G:
            for d in self.G:
                x = self.knn[s]
                y = self.knn[d]
                if (d not in self.cluster[s]) and (len(x & y)>= kt) and (s in y) and (d in x):
                    print("cluster formed")
                    a = self.cluster[s] | self.cluster[d]
                    
                    #Replacing the old cluster with a super cluster
                    for i in a:
                        self.cluster[i] = a 
```

```python
# The main function
g = Graph()
n = int(input("Enter the number of iterations: "))

#The Resultant cluster
res = set()

#Iteration starts
for i in range(n):
    a = set()
    print("--------------------------------------------------\nIteration ",(i+1),"\n\n")
    k,kt = map(int,input("Enter the value for k and kt: ").split(" "))
    g.sNN_cluster(k,kt)
    print("Consolidating Clusters\n")
    
    #Adding the created clusters with length greater than 1 in a set
    for j in g.G:
        if len(g.cluster[j])>1:
            a.add(tuple(g.cluster[j]))
    
    #finding the intersection of the different sets to form a super cluster
    if i == 0:
        res = a
    else:
        c = set()
        for m in a:
            for n in res:
                if len(set(m)&set(n))>1:
                    c.add(tuple(set(m)&set(n)))
                    print("Super cluster formed")
        res = c

#Printing the final cluster
print("\n--------------------------------------------------\nFinal Clusters: ")
for i in res:
    print(i)
```
